```python
import pandas as pd
from datetime import date
```

```python
#https://pypi.org/project/yfinance/
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override() # <== that's all it takes :-)
```

```python
myStocks = pd.DataFrame(columns = ['ticker', 'IPODate'], data = [
    ['TSLA', '2017-01-01'],
    ['APT.AX', '2017-06-26'],
    ['COL.AX', '2018-11-26'],
    ['PGL.AX', '2019-06-11'],
    ['NFLX', '2017-01-01'],
    ['BYND', '2019-05-02'],
    ['SHOP', '2017-01-01'],
    ['EBAY', '2017-01-01'],
    ['PYPL', '2017-01-01'],
    ['BTC-AUD', '2017-01-01']
])
myStocks.head(2)
```

```python
today = date.today().strftime("%Y-%m-%d")
myStockData = pd.DataFrame(columns = ['Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
for index, stock in myStocks.iterrows():
    data = pdr.get_data_yahoo(stock['ticker'], stock['IPODate'], end=today)
    data['Ticker'] = stock['ticker']
    myStockData = pd.concat([myStockData, data], sort=False)

myStockData.reset_index(level=0, drop=False, inplace=True)
myStockData.rename(columns={'index': 'Date'}, inplace=True)
myStockData.head(2)
#myStockData.groupby('Ticker')['Close'].count()
```

## charting

```python
# keep closing price only..
myShortStockData = myStockData[ myStockData['Date']>= '2017-01-01' ]

# pivot df for the chart
myChartData = myShortStockData.pivot(index='Date', columns='Ticker', values='Close')

# removes weekends... using Telsa share data given thats there's a lot of history
myCleanedChartData = myChartData[ pd.notnull(myChartData['TSLA']) ]

myCleanedChartData.head(2)
```

```python
from sklearn import preprocessing
import plotly.graph_objects as go
fig = go.Figure()
for stock in myCleanedChartData.columns:
    fig.add_trace(go.Scatter(x=myCleanedChartData.index, y=preprocessing.scale(myCleanedChartData[stock]), name=stock))

fig.update_layout(
    xaxis_rangeslider_visible=True,
    autosize=True,
    width=1000,
    height=600,
    margin=go.layout.Margin(
        pad=0
    ),
    paper_bgcolor="LightSteelBlue",
)

fig.show()
```
